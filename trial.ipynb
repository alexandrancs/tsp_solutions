```python
def dist(start_place,dirts):
    return (abs(start_place[0] - dirts[0]) + abs(start_place[1] - dirts[1]))
```

```python
db = {}
```

```python
x = 5
y = 5
grid = [(i, j) for j in range(x) for i in range(y)]
```

```python
for i in grid:
    db[i] = [{c:dist(i, c)} for c in grid]
```
